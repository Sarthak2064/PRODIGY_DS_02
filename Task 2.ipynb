```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_df = pd.read_csv('Titanic.csv')

# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Check for missing values
titanic_df.isnull().sum()
```

```python
# For instance, to impute missing ages, you can use the median age:
median_age = titanic_df['Age'].median()
titanic_df['Age'].fillna(median_age, inplace=True)
```

```python
# Now, let's drop the "Cabin" column as it has too many missing values:
titanic_df.drop('Cabin', axis=1, inplace=True)
```

```python
# Lastly, drop rows with missing values in the "Embarked" column:
titanic_df.dropna(subset=['Embarked'], inplace=True)
```

```python
# Distribution of passengers by class
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Relationship between survival and passenger class:
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
# Relationship between survival and age:
sns.histplot(x='Age', hue='Survived', data=titanic_df, bins=30)
plt.title('Survival by Age')
plt.show()
```

```python
# Correlation heatmap:
correlation_matrix = titanic_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Gender distribution among passengers:
sns.countplot(x='Sex', data=titanic_df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Survival rate by gender:
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Fare Distribution by Passenger Class
sns.boxplot(x='Pclass', y='Fare', data=titanic_df)
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
# Age Distribution by Gender
sns.boxplot(x='Sex', y='Age', data=titanic_df)
plt.title('Age Distribution by Gender')
plt.show()
```

```python
# Survival Rate by Siblings/Spouses Aboard (SibSp)
sns.barplot(x='SibSp', y='Survived', data=titanic_df)
plt.title('Survival Rate by Siblings/Spouses Aboard')
plt.show()
```

```python
# Survival Rate by Port of Embarkation (Embarked)
sns.countplot(x='Embarked', hue='Survived', data=titanic_df)
plt.title('Survival by Port of Embarkation')
plt.show()
```

```python
# Age Distribution of Passengers Who Survived
sns.histplot(x='Age', data=titanic_df[titanic_df['Survived'] == 1], bins=30, kde=True, color='green')
plt.title('Age Distribution of Survived Passengers')
plt.show()
```

```python
# Age Distribution of Passengers Who Did Not Survive
sns.histplot(x='Age', data=titanic_df[titanic_df['Survived'] == 0], bins=30, kde=True, color='red')
plt.title('Age Distribution of Non-Survived Passengers')
plt.show()
```
